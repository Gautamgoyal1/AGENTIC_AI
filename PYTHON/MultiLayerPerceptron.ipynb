```python
import math 
import random
```

```python
def sigmoid(x):
    return 1 / (1 + math.exp(-x))
def layer_forward(inputs,weights,biases):
    outputs = []
    for neuron_weights,neuron_bias in zip(weights,biases):
        weighted_sum = sum(i * w for i, w in zip(inputs,neuron_weights)) + neuron_bias
        outputs.append(sigmoid(weighted_sum))
    return outputs
def random_weights(n_neurons,n_inputs):
    return [[random.uniform(-1,1) for _ in range(n_inputs)] for _ in range(n_neurons)]

def random_biases(n_neurons):
    return [random.uniform(-1,1) for _ in range(n_neurons)]
```

```python
# --- Network structure ---
n_inputs  = 3
n_hidden  = 4
n_outputs = 1

# --- Random inputs ---
inputs = [random.uniform(0, 1) for _ in range(n_inputs)]

# --- Random weights and biases ---
hidden_weights = random_weights(n_hidden, n_inputs)
hidden_biases  = random_biases(n_hidden)

output_weights = random_weights(n_outputs, n_hidden)
output_biases  = random_biases(n_outputs)

# --- Forward pass ---
hidden_output = layer_forward(inputs, hidden_weights, hidden_biases)
final_output  = layer_forward(hidden_output, output_weights, output_biases)

# --- Print results ---
print(f"Inputs:        {[round(x, 4) for x in inputs]}")
print(f"Hidden output: {[round(x, 4) for x in hidden_output]}")
print(f"Final output:  {[round(x, 4) for x in final_output]}")
```
