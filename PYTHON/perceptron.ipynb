## Perceptron Class

```python
# Simple perceptron for AND gate

class Perceptron:
    def __init__(self, learning_rate=0.1, epochs=10):
        self.lr = learning_rate
        self.epochs = epochs
        self.weights = [0.21, 0.21]
        self.bias = 0

    def activation(self, z):
        if z >= 0:
            return 1
        return 0
    def activation1(self,z):
        output = 1/(1+np.exp(-z))
        return output
    def predict(self, inputs):
        z = 0
        for i in range(len(inputs)):
            z += inputs[i] * self.weights[i]
        z += self.bias
        return self.activation1(z)

    def train(self, X, y):
        for _ in range(self.epochs):
            for i in range(len(X)):
                prediction = self.predict(X[i])
                error = y[i] - prediction

                # Update weights
                for j in range(len(self.weights)):
                    self.weights[j] += self.lr * error * X[i][j]
                    print(f"Updated weights: {self.weights}")

                # Update bias
                self.bias += self.lr * error


X = [
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
]

y = [0, 0, 0, 1]

p = Perceptron()
p.train(X, y)

print(p.predict([1, 1]))  # 1
print(p.predict([1, 0]))  # 0
```
